```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data
df = pd.read_csv('leaf_shelf_life_data.csv')

# 2. Set the visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 8))

# 3. Create a Correlation Heatmap
# This shows which factors (Temp, Humidity, etc.) have the strongest impact
plt.subplot(2, 2, 1)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation")

# 4. Visualize Exponential Decay (Temp vs Shelf Life)
plt.subplot(2, 2, 2)
sns.scatterplot(data=df, x='Avg_Temp_C', y='Remaining_Shelf_Life', hue='Storage_Type', alpha=0.6)
plt.title("Impact of Temperature on Shelf Life")

# 5. Distribution of Shelf Life
plt.subplot(2, 2, 3)
sns.histplot(df['Remaining_Shelf_Life'], kde=True, color='green')
plt.title("Distribution of Remaining Shelf Life")

# 6. Boxplot: Storage Type Impact
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='Storage_Type', y='Remaining_Shelf_Life')
plt.title("Storage Type: Traditional (0) vs. Controlled (1)")

plt.tight_layout()
plt.show()
```
